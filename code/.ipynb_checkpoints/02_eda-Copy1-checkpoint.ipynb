## Relationship between variables

### Heatmap

```python
plt.figure(figsize= (20, 10))
plt.title('Correlation Plot', fontsize= 35)
sns.heatmap(df.corr(), annot= True);
```

The plot above shows the relationship between the variables. Sales in the dependent variable in my model, and the other variables are independent variables. If the numbers in the matrix are closer to 1, it indicates that the variables have a strong positive linear relationship. It turns out that they is a high linear relationship between customers and sales, which is predictable. If there are more customers, there is a higher chance that stores might have a higher sale. It also turns out that two features, which are open and promo, have strong relationships. Dayofweek shows a negative correlation between sales. Based on the observation in the first notebook, it appears that daysofweek 7 indicates Sundays, but most of stores closed on day 7. Promo2 is highly correlated to promo2sinceweek and promo2sinceyear, which is predictable.

###### School holiday

```python
f= sns.factorplot(data= df,
               x= 'storetype',
               y= 'sales', 
               hue= 'schoolholiday')
f.fig.suptitle('Total Sales Based on School Holidays', fontsize= 30)
f.fig.set_size_inches(10, 10)
f.set_xlabels(fontsize= 20)
f.set_xticklabels(fontsize= 15)
f.set_ylabels('Total Sales', fontsize= 20)
f.set_xticklabels(fontsize= 15)
plt.savefig('../images/sales_holiday');
```

Based on store types and school holiday, there is a difference with regards to sales. In general, store type b has the most highest sales. It might look like there is no difference between sales with regards to the existence of school holidays. For the store types a, c, and d, they tend to have higher sales on school holidays.

```python
plt.figure(figsize= (10, 10))
plt.bar(df[df.schoolholiday == 1].groupby('day').mean()['sales'].index, 
        df[df.schoolholiday == 1].groupby('day').mean()['sales'], color= 'r',alpha= 0.9)
plt.bar(df[df.schoolholiday == 0].groupby('day').mean()['sales'].index, 
        df[df.schoolholiday == 0].groupby('day').mean()['sales'], color= 'r', alpha= 0.3)
plt.title('Average Sales Based on \n School Holidays', fontsize= 30)
plt.xlabel('Day of Week', fontsize= 20)
plt.ylabel('Average Sales', fontsize= 20)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.savefig('../images/barplot_holiday');
```

The thicker red bar shows the average sales on each day when it is school holidays, while the thiner red bar shows the average sales on each day when it is not on school holidays. Generally speaking, the average sales on day of week 3, 4, and 5 are higher when it's on school holidays.

```python
sns.boxplot(x= df.promo,
            y= df.competitiondistance)
```

I wanted to know if the distance between competitors affect the existence of promotion. Based on the plot above, it seems like there is not a much difference. In order to check, we need to run t-test with the null hypothesis being they have the same mean. 

```python
stats.ttest_ind(df[df.promo== 0]['competitiondistance'],
                df[df.promo== 1]['competitiondistance'],
                equal_var= False)
```

It turns out since my p value is larger than 0.05, fail to reject the nill hypothesis, which means that they have the same mean. Therefore, the distance is not a good indicator of the existnece of promotion.
