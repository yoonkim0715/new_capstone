```python
import pandas as pd, numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import scipy.stats as stats
```

### Import the data

```python
df= pd.read_csv('../assets/date_df.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df.set_index(df['date'],inplace=True)
```

## Relationship between variables

### Heatmap

```python
plt.figure(figsize= (20, 10))
plt.title('Correlation Plot', fontsize= 35)
sns.heatmap(df.corr(), annot= True);
```

The plot above shows the relationship between the variables. Sales in the dependent variable in my model, and the other variables are independent variables. If the numbers in the matrix are closer to 1, it indicates that the variables have a strong positive linear relationship. It turns out that they is a high linear relationship between customers and sales, which is predictable. If there are more customers, there is a higher chance that stores might have a higher sale. It also turns out that two features, which are open and promo, have strong relationships. Dayofweek shows a negative correlation between sales. Based on the observation in the first notebook, it appears that daysofweek 7 indicates Sundays, but most of stores closed on day 7. Promo2 is highly correlated to promo2sinceweek and promo2sinceyear, which is predictable.

##### Customers and Sales

```python
for year in ['2013', '2014', '2015']:
    plt.figure(figsize= (8, 8))
    plt.title('Customers & Sales in ' + year , fontsize= 30)
    plt.scatter(df[year]['customers'], df[year]['sales'], color= 'r',alpha= 0.2)
    plt.xticks(fontsize= 15)
    plt.xlabel('Number of Customers', fontsize= 20)
    plt.yticks(fontsize= 15)
    plt.ylabel('Amount of Sales', fontsize= 20)
```

As can be seen and observed from the correlation heatmap, there is a positive linear relationship between customers and sales. Correlation based on years seems to be slightly different with regards to years, but they generally have a strong positive linear relationship. As the number of customers increases, the amount of sales increases as well. 

#### Time and Sales

##### Weekly sales based on years

```python
avg_sales = pd.DataFrame(df.groupby(['year', 'week']).mean()['sales'])
```

```python
avg_sales.reset_index(inplace= True)
```

```python
for year in [2013, 2014, 2015]:
    plt.figure(figsize= (7, 7))
    sns.lineplot(x= 'week', 
                 y= 'sales', 
                 data= avg_sales[avg_sales.year == year])
    plt.title('Weekly Sales in ' + str(year), fontsize= 25)
    plt.xticks(fontsize= 20)
    plt.xlabel('Weeks',fontsize= 18)
    plt.yticks(fontsize= 20)
    plt.ylabel('Sales',fontsize= 18)
    plt.savefig('../images/weekly_sales_in_' + str(year))
```

```python
f= sns.factorplot(x= 'quarter',
              y= 'sales',
              data= df,
              hue= 'year')
f.fig.suptitle('Quarterly Sales Based on Years', fontsize= 18)
f.fig.set_size_inches(12, 9)
f.set_xlabels('Quarters', fontsize= 15)
f.set_xticklabels(fontsize= 12)
f.set_ylabels('Total Sales', fontsize= 15)
f.set_xticklabels(fontsize= 12)
f.savefig('../images/quarters');
```

Similarly, there is a trend with regards to quarters as well. The total sales seem to increase year by year, but not many customers tend to shop at the stores around quarter 1 and 2 and the sales are the highest when quarter 4. Generally speaking, the total sales tend to increase as time passes.

```python
plt.figure(figsize= (8, 8))
plt.title('Sales Based on Days', fontsize= 25)
plt.xlabel('Day of Week', fontsize= 20)
plt.ylabel('Sales', fontsize= 20)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
sns.barplot(x= df.groupby('day').mean().index,
            y= df.groupby('day').mean()['sales'],
            color= 'goldenrod', alpha= 0.4)
plt.savefig('../images/avg_sales_per_day');
```

###### School holiday

```python
f= sns.factorplot(data= df,
               x= 'storetype',
               y= 'sales', 
               hue= 'schoolholiday')
f.fig.suptitle('Total Sales Based on School Holidays', fontsize= 30)
f.fig.set_size_inches(10, 10)
f.set_xlabels(fontsize= 20)
f.set_xticklabels(fontsize= 15)
f.set_ylabels('Total Sales', fontsize= 20)
f.set_xticklabels(fontsize= 15)
plt.savefig('../images/sales_holiday');
```

Based on store types and school holiday, there is a difference with regards to sales. In general, store type b has the most highest sales. It might look like there is no difference between sales with regards to the existence of school holidays. For the store types a, c, and d, they tend to have higher sales on school holidays.

```python
plt.figure(figsize= (10, 10))
plt.bar(df[df.schoolholiday == 1].groupby('day').mean()['sales'].index, 
        df[df.schoolholiday == 1].groupby('day').mean()['sales'], color= 'r',alpha= 0.9)
plt.bar(df[df.schoolholiday == 0].groupby('day').mean()['sales'].index, 
        df[df.schoolholiday == 0].groupby('day').mean()['sales'], color= 'r', alpha= 0.3)
plt.title('Average Sales Based on \n School Holidays', fontsize= 30)
plt.xlabel('Day of Week', fontsize= 20)
plt.ylabel('Average Sales', fontsize= 20)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.savefig('../images/barplot_holiday');
```

The thicker red bar shows the average sales on each day when it is school holidays, while the thiner red bar shows the average sales on each day when it is not on school holidays. Generally speaking, the average sales on day of week 3, 4, and 5 are higher on normal days compared to school holidays.
